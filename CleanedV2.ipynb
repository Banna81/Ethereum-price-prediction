```python
#### Removing Duplicates and Dropping nulls
def clean_df_nulls_dup(df):
    # Check duplicates and drop identical duplicates
    num_duplicates = df.duplicated().sum()
    duplicate_percentage = (num_duplicates / df.shape[0]) * 100
    df = df.drop_duplicates()

    # Check null values and drop columns with over 50% nulls
    null_percentages = df.isnull().sum() / df.shape[0] * 100
    columns_to_drop = null_percentages[null_percentages > 50].index.tolist()
    df = df.drop(columns_to_drop, axis=1)

    # Display datatypes
    data_types = df.dtypes

    # Create cleaning summary
    cleaning_summary = {
        "duplicate_percentage": duplicate_percentage,
        "null_percentages": null_percentages.to_dict(),
        "data_types": data_types.to_dict()
    }

    return df, cleaning_summary
```

```python
def clean_df_string(df, column):
    # Replace commas and similar characters with nothing
    df[column] = df[column].astype(str)
    df[column] = df[column].str.replace(r"[^\d\.]", "")
    try:
        df[column] = pd.to_numeric(df[column])
    except:
        print('couldnot convert')
        pass
    return df
```

```python
#Function that calculate the sum of a column value within a specific date and add to the price data frame in the related row
def merge_totals_by_day(blocks_df, cleaned_price_df, column_to_sum):
    try:
        blocks_df['Date'] = pd.to_datetime(blocks_df['Timestamp (UTC)'])
        cleaned_price_df['Date'] = pd.to_datetime(cleaned_price_df['Date'])

        # Process blocks_df in chunks for memory efficiency
        chunksize = 100000  # Adjust as needed
        totals_by_day = {}
        for chunk_index, block_chunk in enumerate(blocks_df.groupby(pd.Grouper(key='Date', freq='D'))):
            date = block_chunk[0]
            chunk = block_chunk[1]

            print(f"Processing chunk {chunk_index + 1} for date: {date}")  # Debugging output

            total = chunk[column_to_sum].sum()
            totals_by_day[date] = total

        # Create a DataFrame from totals_by_day
        totals_df = pd.DataFrame.from_dict(totals_by_day, orient='index', columns=[f'total {column_to_sum}'])
        totals_df.index.name = 'Date'

        # Merge with cleaned_price_df
        merged_df = cleaned_price_df.merge(totals_df, how='left', on='Date')

        return merged_df

    except Exception as e:
        print(f"An error occurred: {e}")

        return merged_df  
```

```python
#Function that calculate the count of a column value within a specific date and add to the price data frame in the related row
def merge_count_by_day(blocks_df, cleaned_price_df, column_to_count):
    try:
        blocks_df['Date'] = pd.to_datetime(blocks_df['Timestamp (UTC)'])
        cleaned_price_df['Date'] = pd.to_datetime(cleaned_price_df['Date'])

        # Process blocks_df in chunks for memory efficiency
        chunksize = 100000  # Adjust as needed
        totals_by_day = {}
        for chunk_index, block_chunk in enumerate(blocks_df.groupby(pd.Grouper(key='Date', freq='D'))):
            date = block_chunk[0]
            chunk = block_chunk[1]

            print(f"Processing chunk {chunk_index + 1} for date: {date}")  # Debugging output

            total = chunk[column_to_count].nunique()
            totals_by_day[date] = total

        # Create a DataFrame from totals_by_day
        totals_df = pd.DataFrame.from_dict(totals_by_day, orient='index', columns=[f'total {column_to_count}'])
        totals_df.index.name = 'Date'

        # Merge with cleaned_price_df
        merged_df = cleaned_price_df.merge(totals_df, how='left', on='Date')

        return merged_df

    except Exception as e:
        print(f"An error occurred: {e}")
        return None  # Return None to indicate failure
```

```python
def stationary_time_series(df):
    # Candlestick Chart
    fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                                         open=df['Open'],
                                         high=df['High'],
                                         low=df['Low'],
                                         close=df['Close'])])
    fig.update_layout(title='Candlestick Chart',
                      xaxis_title='Date',
                      yaxis_title='Price',
                      xaxis_rangeslider_visible=False)
    fig.show()


    # Augmented Dickey-Fuller Test
    result = adfuller(df['Close'])
    print("Augmented Dickey-Fuller Test:")
    print("ADF Statistic:", result[0])
    print("p-value:", result[1])
    print("Critical Values:")
    for key, value in result[4].items():
        print(f"\t{key}: {value}")
```

```python
def drop_outliers(df, iqr_factor=1.5):
    cleaned_df = df.copy()
    for col in df.columns:
        # Try converting the column to numeric before processing
        try:
            cleaned_df[col] = pd.to_numeric(cleaned_df[col])
        except:
           # Skip non-numeric columns
            continue

        # Calculate interquartile range (IQR)
        q1 = cleaned_df[col].quantile(0.25)
        q3 = cleaned_df[col].quantile(0.75)
        iqr = q3 - q1

        # Define upper and lower bounds for outliers
        upper_bound = q3 + iqr_factor * iqr
        lower_bound = q1 - iqr_factor * iqr

        # Drop rows with values outside the bounds
        cleaned_df = cleaned_df.drop(cleaned_df[col].loc[(cleaned_df[col] > upper_bound) | (cleaned_df[col] < lower_bound)].index)

    return cleaned_df


```

## XGBOOST


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Prepare the data
target_col = 'Close'
predictor_cols = ['Open', 'High', 'Low', 'Adj Close', 'Volume','total Txns','total Block Number']

# Split the data into training and testing sets
train_size = int(len(cleaned1_price) * 0.85)
train_data = cleaned1_price.iloc[:train_size]
test_data = cleaned1_price.iloc[train_size:]
```

```python
# Scale the data
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train_data[predictor_cols + [target_col]])
test_scaled = scaler.transform(test_data[predictor_cols + [target_col]])
```

```python
# Prepare the training and testing data
X_train = train_scaled[:, :-1]
y_train = train_scaled[:, -1]
X_test = test_scaled[:, :-1]
y_test = test_scaled[:, -1]
```

```python
# Build the XGBoost model
model = XGBRegressor()
model.fit(X_train, y_train)

# Make predictions
train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)
```

```python
# Inverse scale the predictions
train_predictions = scaler.inverse_transform(np.concatenate((X_train, train_predictions.reshape(-1, 1)), axis=1))[:, -1]
test_predictions = scaler.inverse_transform(np.concatenate((X_test, test_predictions.reshape(-1, 1)), axis=1))[:, -1]

# Calculate accuracy and error metrics
train_mae = mean_absolute_error(train_data[target_col], train_predictions)
test_mae = mean_absolute_error(test_data[target_col], test_predictions)
train_rmse = np.sqrt(mean_squared_error(train_data[target_col], train_predictions))
test_rmse = np.sqrt(mean_squared_error(test_data[target_col], test_predictions))

print('Train MAE:', train_mae)
print('Test MAE:', test_mae)
print('Train RMSE:', train_rmse)
print('Test RMSE:', test_rmse)
```

```python

X = cleaned_price[['Open', 'High', 'Low', 'Close', 'Adj Close','Volume', 'total Txns', 'total Gas Used', 
         'total Block Number', 'Market Cap', 'Cumulative Return']]
scaler = MinMaxScaler(feature_range=(0, 1))
X = scaler.fit_transform(X)
y = scaler.fit_transform(cleaned_price['Close'].values.reshape(-1,1))

X = X.reshape((X.shape[0], X.shape[1], 1))

train_size = int(len(y) * 0.80)
test_size = len(y) - train_size
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

model = Sequential()
model.add(LSTM(64, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))  

model.compile(loss='mse', optimizer='adam')
model.fit(X_train, y_train, epochs=100, batch_size=16, verbose=1) 

y_pred = model.predict(X_test)
```
